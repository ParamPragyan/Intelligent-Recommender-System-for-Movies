```python
#importing Pandas
import pandas as pd
```

```python
#Read operation on the csv dataset we are using for this Recommender system
df=pd.read_csv('movies.csv')
df.head(5)
#Top 5 rows of the dataset
```

```python
df.shape
```

```python
#To show all columns of the dataset
df.columns
```

```python
#Selecting Features
features=['keywords','cast','genres','director','original_title']
#Replacing all the NaN values with spaces
for feature in features:
    df[feature] = df[feature].fillna('')
    
#Creating a function that will combine all the selected features
def combine_features(row):
    try:
        return row['keywords']+" "+row['cast']+" "+row["genres"]+" "+row["director"]+" "+row['original_title']
    except:
        print ("Error:", row)

df["combined_features"] = df.apply(combine_features,axis=1)
#top 5 rows of newly created column
df["combined_features"].head(5)
```

```python
#Importing all the required packages
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
#creating the count matrix from the new combined column
cv = CountVectorizer()
count_matrix = cv.fit_transform(df["combined_features"])

#computing the cosine similarity based on Count_matrix
cosine_sim = cosine_similarity(count_matrix)
#printing the cosine_sim
cosine_sim.shape
```

```python
def get_title_from_index(index):
    return df[df.index == index]["title"].values[0]

def get_index_from_title(title):
    return df[df.title == title]["index"].values[0]


movie_referred = 'Spectre'

movie_index = get_index_from_title(movie_referred)
```

```python
#printing the index of the specified movie in the  dataset
print(movie_index)
```

```python
i=int(movie_index)
similar_movies = list(enumerate(cosine_sim[i]))
```

```python
similar_movies
```

```python
sorted_similar_movies = sorted(similar_movies,key = lambda x:x[1], reverse = True)

```

```python
sorted_similar_movies
```

```python
i=0
print('The top 10 recommend movies to '+movie_referred+' are: ')
for element in sorted_similar_movies:
    print(get_title_from_index(element[0]))
    i=i+1
    if i>=10:
        break
```
